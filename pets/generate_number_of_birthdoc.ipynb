```python
import random
```

```python
def generate_birth_certificate_number():
    roman_letters = ['I', 'V', 'X', 'L', 'C']
    russian_letters = [chr(code) for code in range(ord('А'), ord('Я') + 1)]

    part1_length = random.randint(1, 4)
    part1 = ''.join(random.choices(roman_letters, k=part1_length))

    part2 = ''.join(random.choices(russian_letters, k=2))
    part3 = f"{random.randint(0, 999999):06d}"

    return f"{part1}-{part2} {part3}"
```

```python
print(generate_birth_certificate_number())
```
